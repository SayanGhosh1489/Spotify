### Removing the NaN values

```python
data.dropna(inplace=True)
data.head()
```

### Changing the album release date to datetime

```python
data['track_album_release_date'] = pd.to_datetime(data['track_album_release_date'])
```

```python
data.info()
```

### Creating new column for year, month and date from track_album_release_date.
### Deleting the column post extraction

```python
data['track_year'] = data['track_album_release_date'].dt.year
data['track_month'] = data['track_album_release_date'].dt.month
data['track_day'] = data['track_album_release_date'].dt.day
data.drop('track_album_release_date', axis=1, inplace= True)
```

```python
data.head()
```

### Univariate and bivariate analysis

```python
#histogram of songs basis year
sns.histplot(data['track_year'], bins = 20)
```

- above graphs shows that most of the released between 2010 and 2019

```python
#mean popular songs vs year
mean_popularity = data.groupby('track_year').agg(track_mean_popularity = ('track_popularity', np.mean))
mean_popularity.sort_values(by = 'track_mean_popularity', ascending= False)
```

```python
plt.figure(figsize = (10,6))
plt.plot(mean_popularity, marker = 'o')
plt.xlabel("Year")
plt.ylabel("Popularity")
plt.title("Year Vs Mean Popularity")
plt.show()
```

#### Ananlysis on popularity

```python
sns.histplot(data['track_popularity'], kde= True, bins = 50,stat='density')
plt.show()
```

### Checking year wise relationship between
- danceability
- energy
- key
- loudness
- mode
- speechiness
- acousticness
- instrumentalness
- liveness
- valence
- tempo
- duration_ms

```python
cols = ['danceability', 'energy', 'key',
       'loudness', 'mode', 'speechiness', 'acousticness', 'instrumentalness',
       'liveness', 'valence', 'tempo','duration_ms']
filtered_df = data.groupby('track_year')[cols].mean()
plt.suptitle("=======Yearwise trend analysis======", fontsize = 16)
for col in cols:
    plt.figure(figsize=(10,6))
    plt.plot(filtered_df.index.values, filtered_df[col], marker ='o')
    plt.title(f"Year wise avg {col} trend",fontsize = 14)
    plt.xlabel('Year',fontsize = 12)
    plt.ylabel(col,fontsize = 12)

plt.show()
```

### Palylist subgenre analysis

```python
data['playlist_subgenre'].unique()
```

```python
if "track_popularity" not in cols:
    cols.append('track_popularity')
playlist_subgenre = data.groupby('playlist_subgenre')[cols].mean()
cols.remove('track_popularity')

for i in range(len(cols)):
    plt.figure(figsize=(10,6))
    playlist_subgenre.sort_values(by=cols[i], ascending= False, inplace= True)
    plt.title(f"Playlist Subgenre Vs Avg {cols[i]}")
    sns.barplot(x = playlist_subgenre.index.values,
               y = playlist_subgenre[cols[i]],palette= 'icefire')
    plt.xticks(rotation = 90)
plt.show()
```

### Yearwise artist, songs analysis
